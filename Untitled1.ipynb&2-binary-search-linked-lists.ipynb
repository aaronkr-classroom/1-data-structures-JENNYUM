```python
def swap(arr, index1, index2):
    # Check if indices are within the bounds of the array
    if index1 < len(arr) and index2 < len(arr):
        temp=arr[index1]
        arr[index1]=arr[index2]
        arr[index2] = temp  # Swap the values
    else:
        print("Error: Index out of bounds.")

# Example usage
array = [10, 20, 30, 40, 50]

print("Before swap:", array)

# Swap values at index 1 and index 3
swap(array, 1, 3)

print("After swap:", array)
```

```python
# Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Example usage for Insertion Sort
array = [12, 11, 13, 5, 6]
print("Original:", array)
insertion_sort(array)
print("after Sorted:", array)

# Function to check if two strings are exactly equal
def compare_strings_exactly(str1, str2):
    return str1 == str2

# Example usage for String Equality
string1 = "hello"
string2 = "hello"

if compare_strings_exactly(string1, string2):
    print(f"{string1} and {string2} are same.")
else:
    print(f"{string1} and {string2} are different.")
```

```python
# Function for Linear Scan to find a value in an array
def linear_scan(arr, value):
    for index, item in enumerate(arr):
        if item == value:
            return index
    return -1

# Example usage
array = [10, 15, 21, 30, 45]
value_to_find = 21
result = linear_scan(array, value_to_find)

if result != -1:
    print(f"Value {value_to_find} found at index {result}.")
else:
    print(f"Value {value_to_find} not found in the array.")
```

```python
# Function for Binary Search to find a value in a sorted array
def binary_search(arr, value):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == value:
            return mid
        elif arr[mid] < value:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Example usage
array = [10, 15, 21, 30, 45]  # Must be sorted for binary search
value_to_find = 21
result = binary_search(array, value_to_find)

if result != -1:
    print(f"Value {value_to_find} found at index {result}.")
else:
    print(f"Value {value_to_find} not found in the array.")
```

```python
# Function to double the size of the array
def array_double(arr):
    new_array = [0] * (len(arr) * 2)  # Create a new array with double the size
    for i in range(len(arr)):
        new_array[i] = arr[i]  # Copy elements to the new array
    return new_array

# Example usage
array = [10, 20, 30, 40, 50]
print("Original array:", array)
new_array = array_double(array)
print("Array after resizing:", new_array)
```

```python
# Define ListNode class
class ListNode:
    def __init__(self, value):
        self.value = value
        self.next = None

# Function to lookup an element in a linked list by index
def linked_list_lookup(head, index):
    current = head
    count = 0
    while current is not None:
        if count == index:
            return current.value
        count += 1
        current = current.next
    return None

# Example usage
node1 = ListNode(10)
node2 = ListNode(20)
node3 = ListNode(30)
node4 = ListNode(40)
node5 = ListNode(50)

# Linking nodes
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Head of the linked list
linked_list_head = node1
element_number = 2
result = linked_list_lookup(linked_list_head, element_number)

if result is not None:
    print(f"Element at index {element_number} is {result}.")
else:
    print(f"Element at index {element_number} does not exist in the linked list.")
```

```python
# Function to insert a node at the end of the linked list
def insert_at_end(head, value):
    new_node = ListNode(value)
    if head is None:
        return new_node
    last = head
    while last.next:
        last = last.next
    last.next = new_node
    return head

# Function to print the linked list
def print_list(head):
    current = head
    while current:
        print(current.value, end=" -> ")
        current = current.next
    print("None")

# Example usage
head = ListNode(10)
head = insert_at_end(head, 20)
head = insert_at_end(head, 30)
head = insert_at_end(head, 40)

print("Original list:")
print_list(head)

# Insert at the beginning
def insert_at_beginning(head, value):
    new_node = ListNode(value)
    new_node.next = head
    return new_node

head = insert_at_beginning(head, 5)
print("After inserting 5 at the beginning:")
print_list(head)

# Insert in the middle
def insert_in_middle(head, value, position):
    if position == 0:
        return insert_at_beginning(head, value)
    new_node = ListNode(value)
    current = head
    for _ in range(position - 1):
        if current is None:
            return head  # Position out of bounds
        current = current.next
    new_node.next = current.next
    current.next = new_node
    return head

head = insert_in_middle(head, 25, 2)
print("After inserting 25 at position 2:")
print_list(head)
```

```python
# Function to delete a node by index
def delete_by_index(head, index):
    if head is None:
        return None
    if index == 0:
        return head.next
    current = head
    for _ in range(index - 1):
        if current is None:
            return head  # Index out of bounds
        current = current.next
    if current.next is None:
        return head  # Index out of bounds
    current.next = current.next.next
    return head

# Function to delete a node by value
def delete_by_value(head, value):
    if head is None:
        return None
    if head.value == value:
        return head.next
    current = head
    while current.next:
        if current.next.value == value:
            current.next = current.next.next
            return head
        current = current.next
    return head

# Example usage
head = ListNode(10)
head = insert_at_end(head, 20)
head = insert_at_end(head, 30)
head = insert_at_end(head, 40)

print("Original list:")
print_list(head)

# Delete by index (e.g., index 2)
head = delete_by_index(head, 2)
print("After deleting node at index 2:")
print_list(head)

# Delete by value (e.g., value 20)
head = delete_by_value(head, 20)
print("After deleting node with value 20:")
print_list(head)
```
